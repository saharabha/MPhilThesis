```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('relative-scores.csv', index_col=0)
df = df.drop(df.loc[df['res'] == 'DM',:].index)
df['res'] = df['res'].astype(int)
df = df.sort_values('res')

plt.figure(figsize=(10,5))
plt.bar(range(len(df)), 
         df['score'])
plt.xticks(range(len(df)), df['res'], rotation = 70)
plt.ylabel('Score relative to A82F/F87V')
plt.xlabel('Residue position')
plt.savefig('bar.png')
plt.show()
```

```python
import numpy as np

plt.set_cmap('coolwarm')

plt.imshow(np.linspace(df['score'].min(), df['score'].max(),100).reshape(-1,1),
          aspect = 0.1)
plt.xticks([])
plt.yticks([1,99], 
           [round(df['score'].min(), 2), 
            round(df['score'].max(), 2)])
plt.savefig('colorbar.png')
plt.show()
```
