```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('relative-scores.csv', index_col=0)
df = df.drop(df.loc[df['res'] == 'DM',:].index)
df['res'] = df['res'].astype(int)
df = df.sort_values('res')

plt.figure(figsize=(5,10))
plt.barh(range(len(df)), 
         df['score'])
plt.yticks(range(len(df)), df['res'])
plt.xlabel('Score relative to A82F/F87V')
plt.ylabel('Residue position')
plt.show()
```
